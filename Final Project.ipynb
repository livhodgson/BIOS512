To start my analysis, I first loaded my packages, tidyverse and RColorBrewer. I then uploaded my data from TidyTuesday, and changed the sex variable from a double to a character. 

```r
library('tidyverse')
library('RColorBrewer')

frogs <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-08-02/frogs.csv')

Sex = as.character(frogs$Female)

frogs$Sex = Sex

head(frogs, show_col_types = FALSE)

options(repr.plot.width=8, repr.plot.height=6)
```

The first plot I made was a bar graph of the types of water where the frogs were located. I further broke down the counts of frogs by sex, and saw that there were more female than male frogs in all locations except for locations with no water.

```r
waterplot = ggplot(frogs, aes(x = Water, fill = Sex)) + geom_bar(position = position_dodge(), color = "darkseagreen4") 
waterplot = waterplot + labs(x = "Water Type", y = "Number of Frogs") + scale_fill_brewer(palette = "GnBu", labels = c('Male', 'Female'))
waterplot
```

I then made another plot similar to the first, but instead of looking at water type, I looked at water location. Similarly to the first graph, I found that females outnumbered males in all locations except for non-aquatic locations.

```r
typeplot = ggplot(frogs, aes(x = Type, fill = Sex)) + geom_bar(position = position_dodge(), color = "darkseagreen4") 
typeplot = typeplot + labs(x = "Location of Water", y = "Number of Frogs") + scale_fill_brewer(palette = "GnBu", labels = c('Male', 'Female'))
typeplot
```

This last graph compares the locations of male and female frogs and the types of vegetation they were found in. Female frogs outnumbered male frogs in all locations except for woody vegetation. 

```r
structureplot = ggplot(frogs, aes(x = Structure, fill = Sex)) + geom_bar(position = position_dodge(), color = "darkseagreen4") 
structureplot = structureplot + labs(x = "Vegetation Type", y = "Number of Frogs") + scale_fill_brewer(palette = "GnBu", labels = c('Male', 'Female'))
structureplot
```

In my analysis, I found that female frogs tended to prefer the aquatic environments as compared with male frogs. Additionally, I found that there were far more female frogs than male frogs. I believe this is because female frogs can only be pregnant with a limited number of eggs at a time, but male frogs can impregnate a large number of female frogs.
